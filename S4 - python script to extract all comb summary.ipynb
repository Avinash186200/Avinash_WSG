```python

for y in year_list:
    
    all_comb = pd.DataFrame()
    
    for file in files:
        
        try:

            new_path = initial_path + file + "/Final Summary/" + str(y) +"/"

            df = pd.read_excel(new_path + "BT_FT summary.xlsx")

            # all_comb = all_comb.append(df,ignore_index = True)
            all_comb = pd.concat([all_comb,df],ignore_index = True)
        
        except Exception as e:
            print(e)
            continue

    all_comb['Accuracy_BT']  = all_comb['Profitable_Trades_x']/all_comb['Total_Trades_x']
    all_comb['PPT_BT'] =  all_comb['Net_P_L_x']/all_comb['Total_Trades_x']

    all_comb = all_comb.rename(columns = {'Accuracy':'Accuracy_FT','PPT':'PPT_FT'})

    
    
    all_comb.to_excel(exp_path + str(y) + " All Stg BT FT summary.xlsx",index = False)

    
    
```
